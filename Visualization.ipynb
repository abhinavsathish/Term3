# Visualization

## Bar Graph

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fruits_name = np.array(['Apple','Banana','Cherry','Dragonfruit'])
costs = np.array([93,45,111,199])
```

```python
plt.bar(fruits_name,costs)
plt.show()
```

```python
student_data = pd.read_excel('/workspaces/Term3/Datasets/Student_Data.xlsx')
```

```python
student_data["Gender"].value_counts().plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Counts")
plt.title('Student Info')
plt.show()
```

## Pie Chart

```python
Assignments = [5,3,1,4]
Labels = ["English","Maths","Computers","Science"]
```

```python
plt.pie(Assignments,labels=Labels,autopct='%1.1f%%')
plt.title("Assignments Pending")
plt.show()
```

## Line Graph

```python
Year = np.arange(2010,2021,1)
AverageRain = np.array([9.8,12,8,7,6,8,9,4,5,8,6.3])
```

```python
plt.plot(Year,AverageRain,color="red",marker='o',markeredgecolor="blue")
plt.title("Average Rainfall")
plt.xlabel("Year")
plt.ylabel("Rainfall")
plt.show()
```

## Double Line Graph

```python
years = [1,2,3]
avg_A = [87,79,99]
avg_B = [91,53,90]
```

```python
plt.plot(years,avg_A,label="Player A")
plt.plot(years,avg_B,"--",label="Player B")
plt.xlabel("Years")
plt.ylabel("Average Runs")
plt.title("Comparision Of Average Runs")
plt.legend()
plt.show()
```
